```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread(r"C:\Users\Admin\CV\messi.jpg")
image
plt.imshow(image)
```

```python
img1 = image.reshape(-1,3)
img1 = np.float32(image)
criteria = (cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,10,2.0)
```

```python
k = 3
attempts = 10
ret,label,center = cv2.kmeans(img1,k,None,criteria,attempts,cv2.KMEANS_RANDOM_CENTERS)
```

```python
res = center[label.flatten()]
res2 = res.reshape(image.shape)
plt.imshow(res2)
```
